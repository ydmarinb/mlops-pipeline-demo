```python
from sklearn.datasets import load_iris



import pandas as pd
from sklearn.datasets import load_iris

# Cargar el conjunto de datos iris
iris = load_iris()
X = iris.data
y = iris.target

# Convertir a DataFrame de pandas y asignar nombres de columnas
column_names = ['feature1', 'feature2', 'feature3', 'feature4']
df = pd.DataFrame(X, columns=column_names)

# Agregar la columna objetivo al DataFrame
df['producto'] = y
```

```python
df['id_usuario'] = range(df.shape[0])
```

```python
df.to_csv('raw_data/feature_data.csv', index=False)
```
